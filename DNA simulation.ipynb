```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating a chromatography graph
x = np.linspace(0, 10, 100)
y = np.exp(-((x - 5) ** 2) / 2) + 0.5 * np.exp(-((x - 7) ** 2) / 2)  # Two peaks
plt.figure()
plt.plot(x, y, linewidth=2)
plt.title('Chromatography Simulation')
plt.xlabel('Retention Time')
plt.ylabel('Intensity')
plt.grid(True)
plt.show()
```

```python
# Simulating a stellar image with random stars
image_size = (100, 100)
num_stars = 50
stellar_image = np.zeros(image_size)

for _ in range(num_stars):
    x = np.random.randint(0, image_size[0])
    y = np.random.randint(0, image_size[1])
    intensity = np.random.randint(100, 256)
    stellar_image[x, y] = intensity

plt.figure()
plt.imshow(stellar_image, cmap='hot')
plt.title('Simulated Stellar Image')
plt.colorbar()
plt.show()
```

This script creates a simulated stellar image by randomly placing 50 bright spots (stars) in a 100x100 pixel image. The stars have random positions and intensities, giving the appearance of a star field. The imshow function displays this image using a heatmap color scheme to make the stars stand out.

Summary
Numpy is used to create and manipulate the image data.
Matplotlib is used to visualize the image.
The image is initially black with zeros, and bright stars are added by setting random pixels to high intensity values.
The image is displayed with a color scheme that highlights the stars.

```python
#DNA sequence simulation 
sequence_length = 100
bases = ['A', 'C', 'G', 'T']
sequence = np.random.choice(bases, sequence_length)
print('DNA Sequence:', ''.join(sequence))

# Simulating DNA sequencing data
dna_sequencing_data = np.random.rand(10, sequence_length)
plt.figure()
plt.imshow(dna_sequencing_data, aspect='auto', cmap='viridis')
plt.title('DNA Sequencing Data')
plt.xlabel('Position in Sequence')
plt.ylabel('Intensity')
plt.colorbar()
plt.show()
```
